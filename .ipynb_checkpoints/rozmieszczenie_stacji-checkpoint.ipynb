```python
stations = data[0]["places"]
```

```python
from ipyleaflet import Map, Marker

center = (data[0]["places_summary"]['latAvg'], data[0]["places_summary"]['lngAvg'])

m = Map(center=center, zoom=11)

for station in stations:
    marker = Marker(location=(station['lat'], station['lng']), draggable=True)
    m.add_layer(marker)
display(m)
```
